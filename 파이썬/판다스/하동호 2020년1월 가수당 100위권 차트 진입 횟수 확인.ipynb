```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import platform
import operator
import matplotlib.font_manager as fm
system_name = platform.system()
if system_name == 'Windows':
# Windows 운영체제
    plt.rc('font', family='Malgun Gothic')
elif system_name == 'Darwin':
# Mac OS
    plt.rc('font', family='AppleGothic')
elif system_name == 'Linux':
# Linux
    path = '/usr/share/fonts/truetype/nanum/NanumMyeongjo.ttf'
    font_name = fm.FontProperties(fname=path, size=12)
    plt.rc('font', family=font_name)
else:
    print("Not support")
```

```python
def graph(self):
    day_rate = self.items()
    x, y = zip(*day_rate)                  #그래프 정의
    plt.figure(figsize=(100, 10))
    plt.plot(x, y)
    plt.xlabel('가수 혹은 그룹 명')
    plt.ylabel('진입횟수')
    plt.title('차트진입횟수')
    plt.show()
```

```python
df = pd.read_csv('KC_music_chart_us_2019.csv')    #파일불러오기, 데이터프레임 작성
```

```python
df['Survey_Base_Date_DE']= pd.to_datetime(df['Survey_Base_Date_DE'], format = '%Y-%m-%d')   #datetime으로 변환
```

```python
df=df.drop(['FILE_NAME','BASE_YMD','Country_CD'], axis=1)        #불필요한 column drop
df  
```

```python
test =[]
test = list(df[df['Survey_Base_Date_DE'].dt.day == 
              int(input('일자를 입력해주세요(7일부터 28일까지입니다) : '))]['Music_NM']) #특정일자 조회
```

```python
day_artist_count={}
for i in test:
    try: day_artist_count[i] += 1
    except: day_artist_count[i]=1        #특정일자의 100위안 랭크 진입한 가수들의 중복 횟수

print(day_artist_count)
```

```python
graph(day_artist_count)         #특정일의 가수당 차트진입횟수
```

```python
test=[]
test= list(df['Music_NM'])   #전체일자조회
```

```python
all_artist_count={}
for i in test:
    try: all_artist_count[i] += 1
    except: all_artist_count[i]=1        #전체 일자의 100위안 랭크 진입한 가수들의 중복 횟수

print(all_artist_count)
```

```python
graph(all_artist_count)                #전체 일의 가수당 차트진입횟수
```

```python
value_list = list(all_artist_count.values())
print(value_list,end =' ')               #가수별 100위권 차트진입횟수 dictionary의 value 값
```

```python
for i in range (len(value_list)):
    value_list[i] = round((value_list[i]/ 22),1)         #모든 Survey_Base_Date_DE의 가수별 100위권 차트진입횟수-새 dictionary의 value 값
    
key_list = list(all_artist_count.keys())                     #새 dictionary 생성을 위해 key리스트 생성
avg_count=dict(zip(key_list,value_list))                 #일 평균 가수별 차트 진입 횟수 -두 리스트를 합쳐서 dictionary생성
```

```python
graph(avg_count)                     # 일일평균 가수당 차트진입횟수
```
